```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import pandas as pd
from bs4 import BeautifulSoup
import time
```

```python
driver = webdriver.Chrome()
```

```python
driver.implicitly_wait(10)
```

```python
driver.get('https://www.linkedin.com/login')
email_input = driver.find_element(By.ID, 'username')
password_input = driver.find_element(By.ID, 'password')
email_input.send_keys('your_email')
password_input.send_keys('your_password')
password_input.send_keys(Keys.ENTER)
```

```python
df=pd.read_csv(r'D:\Website Scrapping\Linkedin_scrapping_1\Excel_files\linkedin_jobs_1p_23_sep.csv')
df
```

```python
data = []
for job_link in (df['Job Link'].iloc[:]):
    driver.get(job_link)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    last_height = driver.execute_script('return document.body.scrollHeight')
    while True:
        driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
        new_height = driver.execute_script('return document.body.scrollHeight')
        if new_height == last_height:
            break
        last_height = new_height

    time.sleep(1) 
    
    try:
        job_title = soup.find('h1', class_='t-24 t-bold job-details-jobs-unified-top-card__job-title').get_text().strip()
    except AttributeError:
        job_title = None

    try:
        job_info = soup.find('div', class_='job-details-jobs-unified-top-card__primary-description').get_text().strip()
    except AttributeError:
        job_info = None
    try:
        job_type = soup.find_all('li', class_='job-details-jobs-unified-top-card__job-insight')[0].find('span').get_text().strip()
    except AttributeError:
        job_type = None
    try:
        employees = soup.find_all('li', class_='job-details-jobs-unified-top-card__job-insight')[1].find('span').get_text().strip()
    except AttributeError:
        employees = None
    try:
        desception = soup.find('div', class_='jobs-description-content__text--stretch').get_text().strip()
    except AttributeError:
        desception = None
    try:
        company_info = soup.find('div', class_='t-14 mt5').get_text().strip()
    except AttributeError:
        company_info = None
    

    data.append({
        'Job Title': job_title,
        'Job Info': job_info,
        'Job Type': job_type,
        'Job Link':job_link,
        'employees':employees,
        'Job Desception':desception,
        'Company Info':company_info
    })
```

```python
df2 = pd.DataFrame(data)
df2.to_csv('linkedin_jobs_full_2_details_1p_23_sep.csv', index=False)
```
